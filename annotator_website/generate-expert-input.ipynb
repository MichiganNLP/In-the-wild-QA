```python
import pandas as pd
import numpy as np
```

```python
def seperate_lins_to_chanels(links):
    link_groups = []
    new_group = []
    for link in links:
        if link is not np.nan:
            new_group.append(link)
        else:
            link_groups.append(new_group)
            new_group = []
    link_groups.append(new_group)
    return link_groups

def expand_links(link_groups,max_len):
    '''
    expand every link group to the max length by randomly sample
    '''
    expanded_groups = []
    for group in link_groups:
        expanded_groups.append(group+list(np.random.choice(group,max_len-len(group))))
    return expanded_groups
```

```python
df = pd.read_csv("../Videos-dropbox-shared-link-HumanSurvival.csv")
links = df["link"]
```

```python
np.random.seed(1)

# seperate links into groups (channels)
link_groups = seperate_lins_to_chanels(links)

# expend to same length
# max_len = max([len(link) for link in link_groups])
# expand_groups = expand_links(link_groups,max_len)
```

```python
# Randomly select videos from the video pool to construct multiples of 5
# randomly select 5 videos for each hit
all_links = [link for group in link_groups for link in group]
all_links = all_links + list(np.random.choice(all_links,(5-len(all_links)%5)%5,replace=False))
np.random.shuffle(all_links)
```

```python
# seperate links into hit (5 each)
input_df = pd.DataFrame(np.array_split(all_links,len(all_links)/5),
                        columns=["video1","video2","video3","video4","video5"])
```
